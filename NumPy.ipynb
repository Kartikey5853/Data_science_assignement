```python
import numpy as np

from sklearn.preprocessing import MinMaxScaler
```

1. Create a NumPy array of random numbers and normalize the data using min-max scaling.

```python
arr = np.random.randint(0,100, size=(3,3),dtype=int)
print(arr)

print("------------------")

scaler = MinMaxScaler()

print(scaler.fit_transform(arr))
```

2. Generate a 5x5 matrix with random integers and replace all occurrences of a specific value
with another.

```python
arr2 = np.random.randint(0,10, size=(5,5),dtype=int)
print(arr2)

print("------------------")

arr2[arr2 == 7] = 12
print(arr2)

```

3. Create two NumPy arrays and perform element-wise addition, subtraction, multiplication, and
division.

```python
arr4 = np.random.randint(1,10, size=(3,3),dtype=int)
print(arr4)

print("------------------")

arr5 = np.random.randint(1,10, size=(3,3),dtype=int)
print(arr5)

print("------------------")

print(arr4 + arr5)

print("------------------")

print(arr4 - arr5)

print("------------------")

print(arr4 * arr5)

print("------------------")

print(arr5 / arr4)

print("------------------")
```

4. Solve the system of linear equations: \(3x + 4y = 7\) and \(5x + 2y = 8\) using NumPy.

```python
x = np.array([[3,4],
              [5,6]])

y = np.array([7,8])

print(np.linalg.solve(x,y))
```

7. Perform matrix multiplication between two 2D arrays.

```python
arr7 = np.random.randint(1,10, size=(3,3),dtype=int)
print(arr7)

print("------------------")

arr8 = np.random.randint(1,10, size=(3,3),dtype=int)
print(arr8)

print("------------------")

print(np.dot(arr7,arr8))

print("------------------")

print(arr7*arr8)
```

8. Calculate the dot product and cross product of two vectors using NumPy.

```python
arr9 = np.random.randint(1,10, size=(3,3),dtype=int)
print(arr9)

print("------------------")

arr10 = np.random.randint(1,10, size=(3,3),dtype=int)
print(arr10)

print("------------------")

print(np.dot(arr9,arr10))

print("------------------")

print(np.cross(arr9,arr10))
```

10. Implement a function using NumPy that returns the inverse of a given matrix.

```python
def inverse(arr):
    return np.linalg.inv(arr)

arr12 = np.random.randint(1,10, size=(3,3),dtype=int)
print(arr12)

print("------------------")

inverse(arr12)
```
