26. Create a 3x3 matrix and find its determinant.

27. Given a 2x2 matrix, compute its inverse and verify by multiplying with the original matrix to
obtain the identity matrix.

```python
arr4 = np.random.randint(0,100, size=(2,2))
print(arr4)
print("-------------------")
arr5 = np.linalg.inv(arr4)
print(arr5)

print("-------------------")

print(np.dot(arr4, arr5 ))
```

28. Calculate the eigenvalues and eigenvectors of a 2x2 matrix.

```python
arr6 = np.random.randint(0,100, size=(2,2))
print(arr6)
print("-------------------")

eigenvalues =   np.linalg.eigvals(arr6)
print("Eigen values",eigenvalues)

print("-------------------")

eigenvectors = np.linalg.eig(arr6)
print("Eigen vectors",eigenvectors)

```

30. Perform Singular Value Decomposition (SVD) on a 3x3 matrix and reconstruct the original
matrix using the SVD components.

```python
arr8 = np.random.randint(0,10, size=(3,3))
print(arr8)
print("-------------------")

U , S , VT = np.linalg.svd(arr8)

sigma = np.diag(S)
print(sigma)

print("-------------------")

reconstruction = U @ sigma @ VT
print(reconstruction)


#NOTE - marking this questions for later revision
```
