using NumPy.
- Normalize a numerical column in a DataFrame using MinMaxScaler from
sklearn.preprocessing and explain the transformation process.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime
```

```python
df = pd.read_csv('used_cars.csv')
```

```python
df = df.drop(columns=['clean_title'])
```

```python
df = df.dropna()
```

```python
for col in df.select_dtypes(include=[np.number]).columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df['model_year_num'] = pd.to_numeric(df['model_year'], errors='coerce')
current_year = datetime.now().year
df['Age of the car'] = current_year - df['model_year_num']
```

```python
scaler = MinMaxScaler()
df['Age_normalized'] = scaler.fit_transform(df[['Age of the car']])
```
