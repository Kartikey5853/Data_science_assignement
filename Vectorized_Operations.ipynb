22. Calculate the dot product of two 1D arrays of size 5.

```python
arr1 = np.random.randint(0,100, size=(1,5))
print(arr1)
print("-------------------")

arr2 = np.random.randint(0,100, size=(1,5))
print(arr2)
print("-------------------")

print(arr1.dot(arr2.T))
```

25. Create a 1D array of 100 random integers and compute the sum of all even numbers using
vectorized operations.

```python
arr6 = np.random.randint(0,100, size=(1,100))
print(arr6)
print("-------------------")

arr6[arr6 % 2 == 0].sum()
```
