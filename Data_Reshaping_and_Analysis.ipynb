11. Data Reshaping and Analysis
- Use NumPy's reshape to change the shape of an array extracted from a DataFrame and analyze the reshaped data.
- Stack two DataFrames vertically and use NumPy to analyze the combined dataset.
- Create a 3D array using NumPy and convert it into a MultiIndex DataFrame, then perform a group operation.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': np.arange(6), 'B': np.arange(10, 16)})
arr = df['A'].to_numpy()
reshaped = arr.reshape(2, 3)
reshaped.mean()
```

```python
df2 = pd.DataFrame({'A': np.arange(6, 12), 'B': np.arange(16, 22)})
stk = np.vstack([df.to_numpy(), df2.to_numpy()])
stk.mean(axis=0)
```

```python
arr3d = np.arange(24).reshape(2, 3, 4)
index = pd.MultiIndex.from_product([[0, 1], [0, 1, 2]], names=['block', 'row'])
df3d = pd.DataFrame(arr3d.reshape(6, 4), index=index)
grouped = df3d.groupby('block').sum()
grouped
```
