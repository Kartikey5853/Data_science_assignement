- Group a DataFrame by a categorical column and calculate the mean and standard deviation of a numerical column for each group.

- Use groupby() to calculate the sum of a column for each group, then apply a NumPy function to the grouped results.

- Create a pivot table from a DataFrame that groups data by two categorical columns and summarizes a numerical column using NumPy operations.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'brand': ['Ford', 'Toyota', 'BMW', 'Ford', 'Toyota', 'BMW', 'Ford', 'Toyota', 'BMW', 'Ford'],
    'type': ['SUV', 'Sedan', 'SUV', 'Sedan', 'SUV', 'Sedan', 'SUV', 'Sedan', 'SUV', 'Sedan'],
    'price': [5000, 7000, 15000, 12000, 9000, 8000, 11000, 10000, 9500, 13000],
    'mileage': [60000, 40000, 30000, 25000, 50000, 35000, 45000, 30000, 20000, 55000]
})
```

```python
grouped = df.groupby('brand')['price'].agg(['mean', 'std'])
grouped
```

```python
sum_mileage = df.groupby('brand')['mileage'].sum()
sqrt_sum = np.sqrt(sum_mileage)
sqrt_sum
```

```python
pivot = pd.pivot_table(df, values='price', index='brand', columns='type', aggfunc=np.mean)
pivot
```
