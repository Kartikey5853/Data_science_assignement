7. Linear Algebra with NumPy
- Solve a system of linear equations using NumPy, and store the solution in a DataFrame.
- Compute the dot product of two columns from a DataFrame using NumPy.
- Perform matrix multiplication on two DataFrames treated as matrices and store the result in a new DataFrame.

```python
import pandas as pd
import numpy as np
```

```python
A = np.array([[2, 1], [1, 3]])
b = np.array([8, 13])
x = np.linalg.solve(A, b)
df_solution = pd.DataFrame({'solution': x})
df_solution
```

```python
df = pd.DataFrame({
    'col1': [1, 2, 3],
    'col2': [4, 5, 6]
})
dot_product = np.dot(df['col1'], df['col2'])
dot_product
```

```python
df1 = pd.DataFrame(np.array([[1, 2], [3, 4]]))
df2 = pd.DataFrame(np.array([[5, 6], [7, 8]]))
matmul_result = np.matmul(df1, df2)
df_matmul = pd.DataFrame(matmul_result)
df_matmul
```
