8. Stack two 3x3 matrices horizontally and vertically.

```python
arr2 = np.random.randint(0,100,size=(3,3))
print(arr2)
arr3 = np.random.randint(0,100,size=(3,3))
print("------------")
print(arr3)
print("------------")
print (np.hstack((arr2,arr3)))

print("------------")
print (np.vstack((arr2,arr3)))
```

9. Concatenate two arrays of different sizes along a new axis.

```python
arr4 = np.random.randint(0,100,size=(3,2))
print(arr4)
print("------------")
arr5 = np.random.randint(0,100,size=(2,3))
print(arr5)  
print("------------")

np.concat((arr4,arr5),axis=1)
```

10. Transpose a 3x2 matrix and then reshape it to have 3 rows and 2 columns.

```python
arr6 = np.random.randint(0,100 , size=(3,2))
print(arr6)
print("------------")

transpose = arr6.transpose()
print(transpose)
print("------------")

arr7 = transpose.reshape(3,2)
print(arr7)
```
