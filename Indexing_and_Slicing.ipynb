11. Given a 1D array of 15 elements, extract elements at positions 2 to 10 with a step of 2.

```python
import numpy as np
```

```python
arr = np.random.randint(0,10, size= (1,15))
print(arr)
print("-------------")
print(arr[0, 1:9:2])
```

13. Replace all elements in a 1D array greater than 10 with the value 10.

```python
arr2 = np.random.randint(0,100, size= (5,5))
print(arr2)
print("-------------")
arr2[arr2 > 10] = 10
print(arr2)

#this is vectorized operations and what it does is loop throught each element 
#after that it assigns a value of 10 if the value is correct 
 
```

15. Create a 1D array of 10 elements and reverse it using slicing.

```python
arr4 = np.random.randint(0,100, size=(1,10))
print(arr4)
print("-------------")
print(arr4[0,::-1])
print("-------------")
```
