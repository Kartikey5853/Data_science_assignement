16. Create a 3x3 matrix and add a 1x3 array to each row using broadcasting.

```python
import numpy as np
```

```python
arr = np.random.randint(0,100, size=(3,3))
print(arr)

print("------------------")

arr1 =  np.random.randint(0,100, size=(1,3))
print(arr)

print("------------------")

result = arr + arr1
print(result)
```

18. Subtract a 3x1 column vector from a 3x3 matrix using broadcasting.

```python
arr3 = np.random.randint(0,100, size=(3,3))
print(arr3)
print("-------------------")
arr4 = np.random.randint(0,100, size=(3,1))
print(arr4)
print("-------------------")

result = arr3 - arr4
print(result)
```
