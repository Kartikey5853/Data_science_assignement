12. Perform groupby operations on a dataset to find the mean and sum of numerical columns
based on a categorical column.

```python
df["price"] = (
    df["price"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
df.groupby("brand")["price"].sum()
```

```python
df.groupby("brand")["price"].mean()
```

14. Merge two DataFrames on a common key and perform an inner, outer, left, and right join.

```python


df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"]
})

df2 = pd.DataFrame({
    "id": [1, 2, 4],
    "score": [85, 90, 95]
})
```

```python
merged_df = pd.merge(df1, df2, on="id", how="outer")
```

```python
merged_df
```

```python
merged_df2 = pd.merge(df1, df2, on="id", how="outer")
merged_df2
```

```python
merged_df3 = pd.merge(df1, df2, on="id", how="left")
merged_df3
```

```python
merged_df4 = pd.merge(df1, df2, on="id", how="right")
merged_df4
```
